```python
import pandas as pd
import numpy as np
import matplotlib_inline as plt


%matplotlib
```

دیتا از :https://www.kaggle.com/datasets/primaryobjects/voicegender/data data:

```python
x=pd.read_csv('voice.csv')
x
```

```python
xx=x[['meanfreq','sd','median','Q25','Q75','IQR','skew','kurt','sp.ent','sfm','mode','centroid','meanfun','minfun','maxfun','meandom','maxdom','dfrange','modindx']]
xx=np.asanyarray(xx)
```

```python
y=x['label'].values
y
```

```python
print(type(y))
```



تبدیل ستون استیرنگ به عدد

Convert the steering column to numeric


```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
x['label']=lb.fit_transform(x['label'])
x['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)

print('set train:',x_train.shape,y_train.shape)
print('set test:',x_test.shape,y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

model=GaussianNB()
model.fit(x_train,y_train)
model
```

```python
y_perd=model.predict(x_test)
y_perd
```



الگوریتم عالی عمل کرده است تمام معیار های ارزیابی 1 شدن و از این بهتر نمیشه

The algorithm performed excellently; all evaluation metrics reached 1, and it doesn't get better than this


```python
from sklearn.metrics import accuracy_score,recall_score,precision_score


print('accuracy:',accuracy_score(y_test,y_perd))
print('recall:',recall_score(y_test,y_perd))
print('precision:',precision_score(y_test,y_perd))
```
