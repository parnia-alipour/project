Stop words کلماتی هستن که زیاد تکرار می‌شن ولی اطلاعات خاصی درباره‌ی معنی یا موضوع متن نمی‌دن، برای همین معمولاً حذف می‌شن



```python
from sklearn.feature_extraction.text import TfidfVectorizer
t=TfidfVectorizer(stop_words='english')
x1=t.fit_transform(file['review'])
x1=x1.toarray()
```

```python
from sklearn.preprocessing import  LabelEncoder
encoder=LabelEncoder()
file['sentiment']=encoder.fit_transform(file['sentiment'])

```

```python
y=file['sentiment'].values
y=np.asanyarray(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y,random_state=4,test_size=0.2)

print('set train:',x_train.shape,y_train.shape)
print('set test:',x_test.shape,y_test.shape)

```

```python
from sklearn import svm
model=svm.SVC()
model.fit(x_train,y_train)

model
```

```python
y_pard=model.predict(x_test)

print(y_pard)
```


The model didn’t receive a good score in this section


نمره دقت ضعیف و پایین است و قابل قبول نیست.

The accuracy score is poor and not acceptable

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
accuracy_score(y_test,y_pard)
```



مدل در این قسمت هم ضعیف عمل کرده است.

The model has also performed poorly in this section.

```python
f1_score(y_test,y_pard,average='macro')
```


مدل در این قسمت هم عملکرد ضعیفی داشته است.

The model has also performed poorly here.

```python
precision_score(y_test,y_pard,average='micro')
```


مدل اصلا در این قسمت امتیاز خوبی نگرفته است.

The model didn’t receive a good score at all in this section

```python
recall_score(y_test,y_pard,average='macro')
```



مدل عملکرد خوبی نداشته است
