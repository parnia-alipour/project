~~~ Data

```python
x=pd.read_csv('marketing_campaign.csv')
x
```

```python
x.isnull().sum()
```

```python
x=x.dropna()
x
```

```python
x=x.drop('ID',axis=1)
x
```

```python
# x=np.asanyarray(x)
```

```python
from sklearn.preprocessing import LabelEncoder

lb=LabelEncoder()
x['Education']=lb.fit_transform(x['Education'])


lb2=LabelEncoder()
x['Marital_Status']=lb2.fit_transform(x['Marital_Status'])
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x=pca.fit_transform(x)
x
```

```python
from sklearn.metrics.pairwise import euclidean_distances
dist_matrix = euclidean_distances(x,x)
print(dist_matrix)
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster_n=4
model=AgglomerativeClustering(linkage='average',n_clusters=cluster_n).fit(x)
print(model.labels_)
```

```python
label_ll=model.labels_
label_ll
```



در این قسمت خوشه بندی نمره خوبی دریافت نکرده است وضعیف عمل کرده است.

"In this clustering section, it has not received a good score and has performed poorly.

```python
from sklearn.metrics import silhouette_score
silhouette_score(x, label_ll)
```


در اینجا هم مدل عملکرد خوبی نداشته و نمره پایینی دریافت کرده.

here too, the model has not performed well and has received a low score.

```python
from sklearn.metrics import davies_bouldin_score

davies_bouldin_score(x, label_ll)
```


برخلاف دو معیار ارزیابی فوق، این مدل در جداسازی خوشه‌ها عملکرد بسیار خوبی داشته و امتیاز خوبی دریافت کرده است.

Unlike the two evaluation metrics above, the model has performed excellently in keeping clusters apart and has received a good score.

```python
from sklearn.metrics import calinski_harabasz_score

calinski_harabasz_score(x, label_ll)
```



بطور کلی , مدل عالی عمل نکرده است و تنها درجداسازی خوشه ها خوب عمل کرده.
